```python
#Importamos las librerias numpy y matplot
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Definimos 2 arreglos de dimensi√≥n 100
ys= 200 + np.random.randn (100)
x = np.arange (0, 100, 1)
```

```python
#Comentario de Prueba 22
#Graficamos los arreglos en un plano 20
plt.plot(x,ys, '.')
plt.fill_between(x, ys, 195, facecolor='c', alpha=0.6)

#Agregamos titulo a la grafica
plt.title ("Sample Visualizacion")

#Agregamos etiquetas a los ejess
plt.xlabel("eje de las x", size =16)
plt.ylabel("eje de las y", size =16)
plt.show()
```
